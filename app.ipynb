```python
import pandas as pd

sales_df = pd.read_csv("monthsales.csv", sep=";")

most_sold = sales_df['Products'].value_counts().idxmax()
least_sold = sales_df["Products"].value_counts().idxmin()
most_amount_sold = sales_df.groupby('Products')['Amount'].sum().max()


print("The least selling produc was:", least_sold)
print("The best-selling product was:", most_sold)
print("Most amount sold was:", most_amount_sold)
display(sales_df)
print(sales_df['Products'].value_counts())
print(sales_df.describe())

```
